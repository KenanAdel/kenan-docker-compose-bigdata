```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('kenan3') \
        .master('spark://spark-master:7077') \
        .getOrCreate()
```

```python
import os

current_path = os.getcwd()
print("Your current location:")
print(current_path)
print("-" * 50)
```

```python
s = "DEST_COUNTRY_NAME STRING , ORIGIN_COUNTRY_NAME STRING,count INTEGER"
```

```python
df_csv = spark.readStream \
    .format("csv") \
    .option("header", "True") \
    .schema(s) \
    .load("file:///home/jovyan/work/csv")

df_read = df_csv.writeStream \
    .format("memory") \
    .queryName("kenan_fresh_stream") \
    .outputMode("append") \
    .start()

df_read.awaitTermination(timeout=5)
```

```python
spark.sql("SELECT * FROM kenan_fresh_stream").show()
```

```python
spark.stop()
```
