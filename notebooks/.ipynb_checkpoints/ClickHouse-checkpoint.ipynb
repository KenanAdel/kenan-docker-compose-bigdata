```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ClickHouseGlobalTest") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

df = spark.createDataFrame([(300, "Clean_Architecture_Success")], ["id", "tag"])

clickhouse_opts = {
    "url": "jdbc:clickhouse://clickhouse:8123/default",
    "user": "clickhouse",
    "password": "clickhouse",
    "driver": "ru.yandex.clickhouse.ClickHouseDriver"
}

df.write.format("jdbc").options(**clickhouse_opts).mode("append").save(dbtable="analytics_ken")
print("Done !!")
```
