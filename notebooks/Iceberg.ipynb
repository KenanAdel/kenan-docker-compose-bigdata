```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ApacheIcebergGlobalTest") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

data = [(10, "Yamani Lakehouse Project", "Yemen")]
columns = ["id", "project_name", "region"]
iceberg_df = spark.createDataFrame(data, columns)

iceberg_df.write \
    .format("iceberg") \
    .mode("append") \
    .saveAsTable("hive_prod.default.ken_iceberg_table")

print("---" * 10)

print("Reading the table:")
spark.read.format("iceberg").load("hive_prod.default.ken_iceberg_table").show()
```
